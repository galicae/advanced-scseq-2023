```python
adata = sc.read_h5ad("/Users/npapadop/Documents/teaching/2023/advanced_scRNAseq/data/chicken_retina/e18_raw.h5ad")
```

```python
fig, ax = plt.subplots()
ax.scatter(np.log(mean), np.log(var), alpha=0.1)
ax.plot(np.arange(-10, 5), np.arange(-10, 5), c="orange")
```
